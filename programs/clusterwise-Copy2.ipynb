```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
# Key Idea - 1
# Train data - 2016
# Test data - 2017
# Area wise clustering - NO
# Independent parameters - Timeslots (morning, noon, evening, night), day of week
# Dependent parameter - average speed of the trip
# Outliers removed using - z score

mean_abs_error_c1 = []
mean_abs_error_c2 = []
mean_abs_error_c3 = []
mean_squared_error_c1 = []
mean_squared_error_c2 = []
mean_squared_error_c3 = []
root_mean_squared_error_c1 = []
root_mean_squared_error_c2 = []
root_mean_squared_error_c3 = []
files_2016 = ["merged_jan.csv", "merged_feb.csv","merged_mar.csv","merged_apr.csv","merged_may.csv","merged_jun.csv", "merged_jul.csv","merged_aug.csv", "merged_sep.csv","merged_oct.csv","merged_nov.csv","merged_dec.csv"]
for index in range(len(files_2016)):
    df = pd.read_csv(files_2016[index], sep=',', usecols=['day','timeslot', 'speed','pickup_dropoff_loc'])  
    z = np.abs(stats.zscore(df))
    df = df[(z < 3).all(axis=1)]
    
    # Seperate the data into clusters
    df_c1 = df.loc[df['pickup_dropoff_loc'] == 1]
    df_c2 = df.loc[df['pickup_dropoff_loc'] == 2]
    df_c3 = df.loc[df['pickup_dropoff_loc'] == 3]
    
    # Get 75% of the rows 
    rows1 = df_c1.shape[0] *75//100
    rows2 = df_c2.shape[0] * 75 // 100
    rows3 = df_c3.shape[0]*75 // 100
    
    # Use 75% of rows for training, and remaining for testing 
    c1_train_data = df.iloc[:rows1]
    c1_test_data = df.iloc[rows1:]
    
    c2_train_data = df.iloc[:rows2]
    c2_test_data = df.iloc[rows2:]
    
    c3_train_data = df.iloc[:rows3]
    c3_test_data = df.iloc[rows3:]
    
    c1_x_train = c1_train_data[['day','timeslot']]
    c1_y_train = c1_train_data['speed']
    
    c2_x_train = c2_train_data[['day','timeslot']]
    c2_y_train = c2_train_data['speed']
    
    c3_x_train = c3_train_data[['day','timeslot']]
    c3_y_train = c3_train_data['speed']
    
    
    
    regresssor_c1 = LinearRegression()
    regresssor_c2 = LinearRegression()
    regresssor_c3 = LinearRegression()
    
    
    regresssor_c1.fit(c1_x_train, c1_y_train)
    regresssor_c2.fit(c2_x_train, c2_y_train)
    regresssor_c3.fit(c3_x_train, c3_y_train)
    
    
    c1_x_test = c1_test_data[['day','timeslot']]
    c1_y_test = c1_test_data['speed']
    
    c2_x_test = c2_test_data[['day','timeslot']]
    c2_y_test = c2_test_data['speed']
    
    c3_x_test = c3_test_data[['day','timeslot']]
    c3_y_test = c3_test_data['speed']
    
    y1_pred = regresssor_c1.predict(c1_x_test)
    y2_pred = regresssor_c2.predict(c2_x_test)
    y3_pred = regresssor_c3.predict(c3_x_test)
    
    mean_abs_error_c1.append(metrics.mean_absolute_error(c1_y_test, y1_pred))
    mean_abs_error_c2.append(metrics.mean_absolute_error(c2_y_test, y2_pred))
    mean_abs_error_c3.append(metrics.mean_absolute_error(c3_y_test, y3_pred))
    
    mean_squared_error_c1.append(metrics.mean_squared_error(c1_y_test, y1_pred))
    mean_squared_error_c2.append(metrics.mean_squared_error(c2_y_test, y2_pred))
    mean_squared_error_c3.append(metrics.mean_squared_error(c3_y_test, y3_pred))
    
    root_mean_squared_error_c1.append(np.sqrt(metrics.mean_squared_error(c1_y_test, y1_pred)))
    root_mean_squared_error_c2.append(np.sqrt(metrics.mean_squared_error(c2_y_test, y2_pred)))
    root_mean_squared_error_c3.append(np.sqrt(metrics.mean_squared_error(c3_y_test, y3_pred)))
    print("*************************************")
    print("For C1 ")
    print('Mean Absolute Error:',metrics.mean_absolute_error(c1_y_test, y1_pred)) 
    print('Mean Squared Error:', metrics.mean_squared_error(c1_y_test, y1_pred))  
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(c1_y_test, y1_pred)))
    
    print("For C2 ")
    print('Mean Absolute Error:', metrics.mean_absolute_error(c2_y_test, y2_pred)) 
    print('Mean Squared Error:', metrics.mean_squared_error(c2_y_test, y2_pred))  
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(c2_y_test, y2_pred)))
    
    print("For C3 ")
    print('Mean Absolute Error:', metrics.mean_absolute_error(c3_y_test, y3_pred)) 
    print('Mean Squared Error:', metrics.mean_squared_error(c3_y_test, y3_pred))  
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(c3_y_test, y3_pred)))
    print("*************************************")
```
