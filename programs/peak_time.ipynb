```python
import numpy as np
from numpy import genfromtxt
import pandas as pd
import collections
from collections import OrderedDict 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./2016_training_data.csv")
x = df['timeslot']
print(df)
```

```python
count_dict = collections.Counter(df['timeslot'])
```

```python
print(count_dict)
```

```python
x = []
y = []
for key in sorted(count_dict):
    x.append(key)
    y.append(count_dict[key])
    print(key,count_dict[key] )
```

```python
font = {'family' : 'Arial',
        'weight' : 'bold',
        'size'   : 15}

plt.rc('font', **font)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x,y, color = 'g')
ax.set_ylabel('Count of trips for year = 2016( y * 10^6)')
ax.set_xlabel('time in 24 hour format')
plt.xticks(x)
plt.show()
```

```python
print("Highest count:", max(y),"Hour :", y.index(max(y))+1)
print("Lowest count: ", min(y), "Hour: ", y.index(min(y))+1)
```
