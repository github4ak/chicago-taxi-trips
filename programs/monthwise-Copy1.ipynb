```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
mean_abs_error = []
mean_squared_error = []
root_mean_squared_error = []
```

```python
# Key Idea - 1
# Train data - 2016
# Test data - 2017
# Area wise clustering - NO
# Independent parameters - Timeslots (morning, noon, evening, night), day of week
# Dependent parameter - average speed of the trip
# Outliers removed using - z score
mean_abs_error = []
mean_squared_error = []
root_mean_squared_error = []



# df_1 = pd.read_csv("short_trips_2016.csv", sep=',', usecols=['day','timeslot', 'speed'])  
# df_2 = pd.read_csv("long_trips_2016.csv", sep=',', usecols=['day','timeslot', 'speed'])  
# df_1['duration'] = 10
# df_2['duration'] = 11

# frames = [df_1, df_2]
df = pd.read_csv("short_trips_2016.csv", sep=',', usecols=['day','timeslot', 'speed'])  
df['duration'] = 10
# print(df)

# z = np.abs(stats.zscore(df))
# df = df[(z < 3).all(axis=1)]

X = df[['day','timeslot','duration']]
Y = df['speed'] 
# print(X)
# print(Y)
regressor = LinearRegression()  
regressor.fit(X, Y)

df_2017 = pd.read_csv("short_trip_2017.csv", sep=',', usecols=['day','timeslot', 'speed'])  
df_2017['duration'] = 10
# print(df_2017)
# z = np.abs(stats.zscore(df_2017))


# df_1_17 = pd.read_csv("short_trip_2017.csv", sep=',', usecols=['day','timeslot', 'speed'])  
# df_2_17 = pd.read_csv("long_trips_2017.csv", sep=',', usecols=['day','timeslot', 'speed'])  
# df_1_17['duration'] = 10
# df_2_17['duration'] = 11

# frames_1 = [df_1_17, df_2_17]

# df_2017 = pd.concat(frames_1)




X_2017 = df_2017[['day','timeslot','duration']]
y_2017 = df_2017['speed'] 
y_pred = regressor.predict(X_2017)

print('Mean Absolute Error:', metrics.mean_absolute_error(y_2017, y_pred)) 
mean_abs_error.append(metrics.mean_absolute_error(y_2017, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_2017, y_pred))  
mean_squared_error.append(metrics.mean_squared_error(y_2017, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_2017, y_pred)))
root_mean_squared_error.append(np.sqrt(metrics.mean_squared_error(y_2017, y_pred)))
```

```python
print('Average Mean Absolute Error:',sum(mean_abs_error)/len(mean_abs_error))
print("Average Mean Squared Error:", sum(mean_squared_error)/len(mean_squared_error))
print('Average Root Mean Squared Error:', sum(root_mean_squared_error)/len(root_mean_squared_error))
```
